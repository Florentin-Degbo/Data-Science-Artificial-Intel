**What is EDA (Exploratory Data Analysis)**
1. EDA is a process in data analytics used to analyze and investigae datasets and find out the main characteristics,
2. EDA helps data analyst to discover patterns in the dataset and examine their assumptions,
3. EDA is also used to find out about the relationship btween the dataSET VARIABLES,
4. It was developed by the Americsn mathematician, John Tukey, in the 190s,
5. EDA helps analysts to answer the management team questions,
6. When EDA is done, the dataset can be used for more sophisticated application of AI such as ML.

**Type of EDA:**
1. Univariate: Examining one variable
2. Bivariate: Examining two variables
3. Multivariate: Examining more than two variables

**EDA Tools:**
1. Graphical EDA (data vizualization): plots, graphs, charts
2. Nongraphical EDA: Examining the relationships betweeen the dataset variables.

**EDA Most Popular Languages:**
1. Pyhton
2. R

### Data Indexers
- There are two types of indexers:
1. iloc indexer: This is implicit indexing (not clearly described). It is interger-based and refers to the position of the element. I stars with 0. 
2. loc indexer: This is explicit indexing (clearly stated). It is the user-defined name and acts as a locator of the element. 

```python
# 2. What is the movie sitting in position 300?
movies.iloc[300]
```

```python
# 3. What is the information about the movie "Forrest Gump"?
movies.loc['Forrest Gump']
```

```python
# 4. Is there only one version of the movie "101 Dalmatians"
movies.loc['101 Dalmatians']
```
