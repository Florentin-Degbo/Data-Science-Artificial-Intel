### Introduction

**What is data visualization (Viz)**
- Graphical presentation of data.

**What are the benefits of data visualization (Viz)?**
1. Helps with discoverring the data patterns,
2. Helps with more effective communication,
3. Helps non-technical people to understand the data beter or easier,
4. Helps with understanding the data distribution,
5. Helps with discovering the current trends in te dataset.

**What are the types of data visualization?**
There are 3 types of data visualizaion:
1. Univariate Visualization: plotting one variable
2. Bivariate Visualization: plotting two variable
3. Multivariate Visualization: plotting more than two variables

**What are popular types of data visualization plots?**
1. Line Plos
2. Bar Plots
3. Pie Charts
4. Scatter Plots
5. Histograms
6. Heatmaps

**What are popular libairies used for data visualization (Viz)**
1. Matplotlib.pyplot
2. seaborn


```python
# importing librairies

# general librairies
import numpy as np
import pandas as pd
import math 

# data visualization librairies
import matplotlib.pyplot as plt
import seaborn as sns 


# to suppress warnings librar
import warnings 
warnings.filterwarnings('ignore')

# 1. These Python libraries for data visualization are said to be platform-independence:
# this means that we can use these librairies on any OS: windows, linux and IOS
# 2. To enhance the performance of matplotlib, an API has been added to it. 
# This API is pyplot.
# * API: (Application Programming Interface): Acts as a bridge that connects two or more applications. 
```

```python
# 1. line plots: these are the simplest plots
# they display the progress of an event over time

# creating the dataset
a = np.linspace(0, 10)

# plotting the dataset: the sin and cos of the number 0-10
plt.plot(a, np.sin(a)) # the sin of 0 is 0
plt.plot(a, np.cos(a)); # the cos of o is 1
```

```python
# creating the dataset
x = np.array([0, 5])
y = np.array([0, 20])

# plotting the dataset: 
plt.plot(x, y, color='red'); 
```

```python
# creating the dataset
y = np.array([12, 0, 9, 2, 14])

# plotting the dataset: 
plt.plot(y, color='green', linestyle='dashed', marker = "*"); 
```

```python
# creating the dataset
x = pd.Series(['Sunday', 'Monday', 'Tuesday', 'Thursday'])
y = pd.Series(['Running', 'Tennis', 'Football', 'Biking'])

# plotting the dataset: 
plt.plot(x,y)
plt.title('Days vs Sports')
plt.xlabel('Days')
plt.ylabel('Sports')
plt.grid();
```

```python
# 2. bar plots: these are good to compare various categories


# creating the dataset
y = np.array([5, 8, 10, 2])
x = np.array(['Five', 'Eight', 'Ten', 'Two'])

# plotting the dataset: 
plt.bar(x,y, color='r')
plt.title('Numbers vs Written Forms');
```

```python
# 3. pie chars: these are good to show the proportion over the whole

# creating the dataset
y = np.array([5, 20, 10, 8, 25])
cars = ['Benz', 'Ford', 'Honda', 'Tesla', 'Maserati']
spaces = [0, 0.2, 0, 0, 0.1]
kolrs = ['b', 'c', 'm', 'y', 'g']

# plotting the dataset: 
plt.pie(y, labels = cars, explode = spaces, colors = kolrs) 
plt.legend();
```

**Color Codes**

1. r for red
2. g for green
3. b for blue
4. c for cyan
5. m for magenta
6. y for yellow
7. k for black
8. w for white

**Heatmaps**

1. heatmaps are seaborn librairy plots that display correlation, 
2. what is correlation? Correlation the relationship between variables ,
3. correlation can be positive or negative,
4. if the variables move together in either directions (up or down), we call this a positive correlation,
5. if the variables move in opposite directions (one goes up and the other goes down), we call this a negative correlation
6. a heatmap is designed for numeric values only.
7. the range of corelation in the heatmap goes from -1 to +1

```python
# let's create a new dataset and name it num_df that contains only numeric variables of titanic
num_df = titanic[['Age', 'Fare', 'Survived', 'Pclass', 'SibSp', 'Parch']]
num_df
```

```python
# correlation table
num_df.corr()
```

**Observations-5**:

-Add 5 points
1. SibSp and Parch exhibit a positive correlation (0.414838), indicating that passengers with siblings/spouses often traveled with parents/children as well.
2. Age has weak correlations with most other variables, showing that age didn't strongly influence survival, fare, or family relationships.
3. Age has a moderate negative correlation with Pclass (-0.369226), implying that older passengers were more likely to travel in lower classes
4. Survived has a positive correlation with Fare (0.257307), suggesting that passengers who paid higher fares had a better chance of survival.
5. Pclass shows a strong negative correlation with Fare (-0.549500), indicating that higher-class passengers (lower Pclass values) paid higher fares.

```python
# let's create the heatmap for this dataset
plt.figure(figsize = (12, 8)) # setting the size of the heatmap

sns.heatmap(num_df.corr(), annot=True, cmap= 'RdYlGn')
```
